```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://data.cdc.gov/api/views/r8kw-7aab/rows.csv?date=20231105&accessType=DOWNLOAD')
df
```

```python
df = df.drop(columns = 'Footnote')
```

```python
df
```

```python
df.drop_duplicates()
```

```python
sum_of_cols = df.sum(axis=0)
print(sum_of_cols)
```

```python
# Calculate the sum of the specific column
sum_of_column = df['Pneumonia, Influenza, or COVID-19 Deaths'].sum()
print(sum_of_column)
```

```python
new_column_name = "Weekly Tracking"  #Replace with new column name
df = df.rename(columns={"MMWR Week": new_column_name})
df
```

```python
new_column_name = "Recorded Date"  #Replace with new column name
df = df.rename(columns={"Data as of": new_column_name})
df
```

```python
# List of column names to convert to datetime datatype
date_columns = ['Start Date', 'End Date', 'Week Ending Date']

# Converting the specified columns to datetime
df[date_columns] = df[date_columns].apply(pd.to_datetime)
```
